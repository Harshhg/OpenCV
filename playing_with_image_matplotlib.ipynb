```
# Importing cv2 module.. package name -> (opencv-contrib-python)
import cv2
import matplotlib.pyplot as plt
```

```
# Checking version
print(cv2.__version__)
```

```
# Reading Image  in array
img=cv2.imread('dog.jpg',1)
#  1  -  means image in same color
#  0  -  means image in Black and White  --- Gray Image
# -1  -  maintain image transparency (in case of image over image)

# To split the image into BGR
b,g,r=cv2.split(img)
```

```
# Printing the shape of the image
print(img.shape)  
# Height, width, color channel(RGB/BGR) (0-means black and white)
# In openCV  the color channel order is BGR.
# In GAN the color channel order is RGB
```

```
# Using matplotlib to display the image
plt.imshow(img)  # Matplotlib will display the image in RGB format...(but the image is in BGR..)
```

```
# Now we will reverse the BGR to RGB ..
img=cv2.merge([r,g,b])
plt.imshow(img)  # Now it will display the original image(only in case of matplotlib).

# Removing the x-axis and y-axis parameters..
plt.xticks([])
plt.yticks([])
```

```
# Displaying the B format (from BGR) that we splitted..
plt.imshow(b)
```

```
# Displaying the G format (from BGR) that we splitted..
plt.imshow(g)
```

```
# Cropping the image from numpy array
plt.imshow(img[100:300,200:400])  # Cropped image
```

```
# Drawing a line on the image..
img=cv2.line(img,(0,0),(300,300), (0,0,255),3)
# arguments -> (image, (starting position), (ending position), (color in RGB format), thickness of line)

# Drawing a rectangle..
img=cv2.rectangle(img,(400,100),(200,250),(0,255,0),3)

# Drawing a circle..
img=cv2.circle(img, (500, 300), 100, (255, 0, 0), 2)
# arguments -> (image, (center point), radius, (color in RGB), thickness)

plt.imshow(img)
```

```
# To save/write the image after editing
cv2.imwrite('newdog.jpg',img)
```
